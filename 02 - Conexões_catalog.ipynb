```python
import json
from google.oauth2 import service_account
import google.auth.transport.requests

# Caminho do arquivo JSON da Service Account no Workspace
SERVICE_ACCOUNT_FILE = "/Workspace/fraude-e-comerce/chaves/unitycatalog-479702-2fc49a967079.json"

# Carregar credenciais a partir do arquivo
creds = service_account.Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE,
    scopes=["https://www.googleapis.com/auth/cloud-platform"]
)

# Atualizar credenciais para obter o token
request = google.auth.transport.requests.Request()
creds.refresh(request)

# Bearer Token pronto para uso
token = creds.token
print("Bearer Token:", token)
```

```python
from databricks.sdk import WorkspaceClient
import requests

w = WorkspaceClient()
conn = w.connections.get("gcp_cloudbuild")

host = conn.options["host"]
token = conn.options.get("bearer_token")  # usar .get() evita erro se não existir

headers = {"Authorization": f"Bearer {token}"}
url = f"{host}/v1/projects/unitycatalog-479702/builds"

resp = requests.get(url, headers=headers)
print(resp.json())
```
